```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
random_state = 42

%matplotlib inline
```

## Feature Selection

```python
correlation_matrix = train_df.fillna(train_df.mode(0).iloc[0]).corr()
feature = []
for idx, value in enumerate(correlation_matrix['Target'][1:]):
    if np.abs(value) > 0.15:
        feature.append('X'+str(idx+1))
        print(value)
print(feature)
```

```python
features = feature + kk
```

```python
train_df_feature = train_df[features + ['Target']]
```

# 모델링

```python
from sklearn.metrics import confusion_matrix

def custom_score(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    return 300*fn + 10*fp
```

```python
def custom_metric(y_true, y_pred):
    y_pred = list(map(lambda x: x>= 0.5, y_pred))
    metric_name = 'custom_metric'
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    value = 300*fn + 10*fp
    
    is_higher_better = False
    return metric_name, value, is_higher_better
```

```python
def custom_score_deep(y_true, y_pred):

    y_pred = list(map(lambda x: x>= 0.5, y_pred))
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    return 300*fn + 10*fp
```

```python
def custom_score_xgb(y_pred, y_true):
    y_true = y_true.get_label()
    y_pred = list(map(lambda x: x>= 0.5, y_pred))
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    
    return 'custom_score', 300*fn + 10*fp
```

## XGB with Optuna

```python
import optuna 
from optuna import Trial, visualization
from optuna.samplers import TPESampler
from imblearn.combine import SMOTETomek
from imblearn.under_sampling import TomekLinks

import xgboost as xgb
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import StratifiedKFold
from imblearn.over_sampling import ADASYN

def custom_score_xgb(y_pred, y_true):
    y_true = y_true.get_label()
    y_pred = list(map(lambda x: x>= 0.5, y_pred))
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    
    return 'custom_score', 300*fn + 10*fp


X = train_df_feature.drop('Target', axis=1)
y = train_df_feature.Target

X = X.fillna(X.mode().iloc[0])

def objectiveXGB(trial: Trial, X, y, test):
    param = {
        "n_estimators" : trial.suggest_int('n_estimators', 500, 4000),
        'max_depth':trial.suggest_int('max_depth', 8, 16),
        'min_child_weight':trial.suggest_int('min_child_weight', 1, 300),
        'gamma':trial.suggest_int('gamma', 1, 3),
        'learning_rate': 0.01,
        'colsample_bytree':trial.suggest_discrete_uniform('colsample_bytree',0.5, 1, 0.1),
        'nthread' : -1,
        'lambda': trial.suggest_loguniform('lambda', 1e-3, 10.0),
        'alpha': trial.suggest_loguniform('alpha', 1e-3, 10.0),
        'subsample': trial.suggest_categorical('subsample', [0.6,0.7,0.8,1.0] ),
        'random_state': random_state
    }
    ## XGBRegressor를 예시로 사용하였습니다.

    skf = StratifiedKFold(n_splits = 4, shuffle = True, random_state = random_state)
    n_iter = 0
    
    featuress = X
    label = y

    s_xgb = []
    
    for train_idx, test_idx in skf.split(featuress, label):
    
        n_iter += 1
        print(f'--------------------{n_iter}번째 KFold-------------------')
        print(f'train_idx_len : {len(train_idx)} / test_idx_len : {len(test_idx)}')
        
        label_train = label.iloc[train_idx]
        label_valid = label.iloc[test_idx]
        
        X_train_cv, X_valid_cv = featuress.iloc[train_idx, :], featuress.iloc[test_idx, :]
        y_train_cv, y_valid_cv = label.iloc[train_idx], label.iloc[test_idx]
    
        X_train_cv, y_train_cv = SMOTETomek(tomek=TomekLinks(sampling_strategy='majority')).fit_resample(X_train_cv, y_train_cv)
        
        ### 여기서부터 모델 생성
        ### XGB
        xgb_classifier = xgb.XGBClassifier(**param)
        xgb_classifier.fit(
            X_train_cv, y_train_cv,
            eval_set=[(X_valid_cv, y_valid_cv)],  # 유효성 검사 데이터셋
            eval_metric=custom_score_xgb,          # 평가 메트릭
            callbacks=[xgb.callback.EarlyStopping(10)]
        )
        y_pred_cv_xgb = xgb_classifier.predict(X_valid_cv)
        
        s_xgb.append(custom_score(y_valid_cv, y_pred_cv_xgb))
    
    
    score = sum(s_xgb)/len(s_xgb)

    return score
```

```python
# direction : score 값을 최대 또는 최소로 하는 방향으로 지정 
study = optuna.create_study(direction='minimize',sampler=TPESampler())

# n_trials : 시도 횟수 (미 입력시 Key interrupt가 있을 때까지 무한 반복)
study.optimize(lambda trial : objectiveXGB(trial, X,  y, []), n_trials=500)
print('Best trial: score {},\nparams {}'.format(study.best_trial.value,study.best_trial.params))
```

## LGBM with Optuna

```python
import optuna 
from optuna import Trial, visualization
from optuna.samplers import TPESampler

import lightgbm as lgb
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import StratifiedKFold

X = train_df_feature.drop('Target', axis=1)
y = train_df_feature.Target

X = X.fillna(X.mode().iloc[0])

def objectiveLGB(trial: Trial, X, y, test):
    param = {
        'objective': 'regression', # 회귀
        'verbose': -1,
        'metric': 'rmse', 
        'max_depth': trial.suggest_int('max_depth',3, 15),
        'learning_rate': trial.suggest_loguniform("learning_rate", 1e-8, 1e-2),
        'n_estimators': trial.suggest_int('n_estimators', 100, 3000),
        'min_child_samples': trial.suggest_int('min_child_samples', 5, 100),
        'subsample': trial.suggest_loguniform('subsample', 0.4, 1),
    }

    skf = StratifiedKFold(n_splits = 4, shuffle = True, random_state = random_state)
    n_iter = 0
    
    featuress = X
    label = y

    s_lgb = []
    
    for train_idx, test_idx in skf.split(featuress, label):
    
        n_iter += 1
        print(f'--------------------{n_iter}번째 KFold-------------------')
        print(f'train_idx_len : {len(train_idx)} / test_idx_len : {len(test_idx)}')
        
        label_train = label.iloc[train_idx]
        label_valid = label.iloc[test_idx]
        
        X_train_cv, X_valid_cv = featuress.iloc[train_idx, :], featuress.iloc[test_idx, :]
        y_train_cv, y_valid_cv = label.iloc[train_idx], label.iloc[test_idx]
    
        X_train_cv, y_train_cv = SMOTETomek(tomek=TomekLinks(sampling_strategy='majority')).fit_resample(X_train_cv, y_train_cv)
        
        ### 여기서부터 모델 생성
        ### XGB
        lgb_classifier = lgb.LGBMClassifier(**param)
        lgb_classifier.fit(
            X_train_cv, y_train_cv,
            eval_set=[(X_valid_cv, y_valid_cv)],  # 유효성 검사 데이터셋
            eval_metric=custom_metric,         # 평가 메트릭
            callbacks=[lgb.early_stopping(10, first_metric_only=False, verbose=True, min_delta=0.0)]
        )
        y_pred_cv_lgb = lgb_classifier.predict(X_valid_cv)
        
        s_lgb.append(custom_score(y_valid_cv, y_pred_cv_lgb))
    
    
    score = sum(s_lgb)/len(s_lgb)
    print(s_lgb)

    return score
```

```python
# direction : score 값을 최대 또는 최소로 하는 방향으로 지정 
lgbmstudy = optuna.create_study(direction='minimize',sampler=TPESampler(seed = random_state))

# n_trials : 시도 횟수 (미 입력시 Key interrupt가 있을 때까지 무한 반복)
lgbmstudy.optimize(lambda trial : objectiveLGB(trial, X,  y, []), n_trials=500)
print('Best trial: score {},\nparams {}'.format(lgbmstudy.best_trial.value,lgbmstudy.best_trial.params))
```

```python
xgb_params = {'n_estimators': 654, 'max_depth': 12, 'min_child_weight': 11, 'gamma': 1, 'colsample_bytree': 0.5, 'lambda': 4.31742702882685, 'alpha': 0.00989655020027596, 'subsample': 0.8}
lgbm_params = {'max_depth': 15,
 'learning_rate': 0.00012235676716719124,
 'n_estimators': 476,
 'min_child_samples': 84,
 'subsample': 0.665433990013184}
```

## Full Model for 제출

```python
X_test = pd.read_csv('test.csv')
sample_submission = pd.read_csv('sample_submission.csv')

for col in cols_to_change:
    X_test[col] = X_test[col].apply(lambda x: 0 if x < 2 else 1)

anomal = []
for idx, i in enumerate((X_test.max() > 10).to_list()):
    if i:
        anomal.append(X_test.columns[idx])
```

```python
from sklearn.model_selection import StratifiedKFold
from keras.callbacks import EarlyStopping
import xgboost as xgb
from imblearn.combine import SMOTETomek
from imblearn.under_sampling import TomekLinks
import lightgbm as lgb

import tensorflow as tf
import random as rn
import time

np.random.seed(random_state)
tf.random.set_seed(random_state)
rn.seed(random_state)
xgb.set_config(verbosity=0)

start = time.time()

X = train_df_feature.drop('Target', axis=1)
y = train_df_feature.Target

X = X.fillna(X.mode().iloc[0])

skf = StratifiedKFold(n_splits = 4, shuffle = True, random_state = random_state)
n_iter = 0

featuress = X
label = y

score_list = []
s_xgb = []
s_lgb = []
s_deep = []
s_ext = []
middle = []
end = []

y_pred_all = []
y_valid_all = []

for train_idx, test_idx in skf.split(featuress, label):
    
    n_iter += 1
    print(f'--------------------{n_iter}번째 KFold-------------------')
    print(f'train_idx_len : {len(train_idx)} / test_idx_len : {len(test_idx)}')
    
    label_train = label.iloc[train_idx]
    label_valid = label.iloc[test_idx]
    
    X_train_cv, X_valid_cv = featuress.iloc[train_idx, :], featuress.iloc[test_idx, :]
    y_train_cv, y_valid_cv = label.iloc[train_idx], label.iloc[test_idx]

    X_train_cv, y_train_cv = SMOTETomek(tomek=TomekLinks(sampling_strategy='majority')).fit_resample(X_train_cv, y_train_cv)
    
    ### 여기서부터 모델 생성

    ### XGB
    xgb_classifier = xgb.XGBClassifier(**xgb_params)
    xgb_classifier.fit(
        X_train_cv, y_train_cv,
        eval_set=[(X_valid_cv, y_valid_cv)],  # 유효성 검사 데이터셋
        eval_metric=custom_score_xgb,          # 평가 메트릭
        callbacks=[xgb.callback.EarlyStopping(10)]
    )
    y_pred_cv_xgb = xgb_classifier.predict(X_valid_cv)
    
    s_xgb.append(custom_score(y_valid_cv, y_pred_cv_xgb))
    
    ### LGBM
    lgb_classifier = lgb.LGBMClassifier(**lgbm_params)
    lgb_classifier.fit(
        X_train_cv, y_train_cv,
        eval_set=[(X_valid_cv, y_valid_cv)],  # 유효성 검사 데이터셋
        eval_metric=custom_metric,         # 평가 메트릭
        callbacks=[lgb.early_stopping(10, first_metric_only=False, verbose=True, min_delta=0.0)]
    )
    y_pred_cv_lgb = lgb_classifier.predict(X_valid_cv)
    s_lgb.append(custom_score(y_valid_cv, y_pred_cv_lgb))
    
    
    ### Deep

    early_stopping = EarlyStopping(monitor='val_loss', patience=20)
    
    deep_model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(units=16, activation="relu", input_shape=(X_train_cv.shape[-1],) ),
        tf.keras.layers.Dense(units=1, activation="sigmoid"),
    ])
    deep_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['Accuracy'])

    history = deep_model.fit(X_train_cv, y_train_cv, epochs=200, validation_split=0.25, callbacks=[early_stopping], verbose=0, class_weight= {0: 1/4, 1: 3/4})
    
    y_pred_cv_deep = deep_model.predict(X_valid_cv)
    y_pred_cv_deep = list(map(lambda x: x>= 0.5, y_pred_cv_deep))
    deeps = custom_score(y_valid_cv, y_pred_cv_deep)
    print('deep score:', deeps)
    s_deep.append(deeps)

    ### Voting
    y_pred_cv = [(x + y + z)/3 for x, y, z in zip(y_pred_cv_xgb, y_pred_cv_lgb, y_pred_cv_deep)]
    y_pred_cv = list(map(lambda x: 1 if x>= 0.5 else 0, y_pred_cv))
    
    score = custom_score(y_valid_cv, y_pred_cv)
    
    print(f'{n_iter}번째 단일 score:{score}')
    score_list.append(score)

    middle.append(time.time())

    ## Prediction
    X_testt = X_test.fillna(train_df.mode().iloc[0])[features] # data leakage 조심!
    
    
    y_test_pred_xgb = xgb_classifier.predict(X_testt)
    y_test_pred_lgb = lgb_classifier.predict(X_testt)
    
    y_test_pred_deep = deep_model.predict(X_testt)
    y_test_pred_deep = list(map(lambda x: 1 if x>= 0.5 else 0, y_test_pred_deep))
    
    y_test_pred = [(x + y + z)/3 for x, y, z in zip(y_test_pred_deep, y_test_pred_xgb, y_test_pred_lgb)]
    y_test_pred = list(map(lambda x: 1 if x>= 0.5 else 0, y_test_pred))
    
    if not y_pred_all:
        y_pred_all = y_test_pred
    else:
        y_pred_all = [(x + y) for x, y in zip(y_test_pred, y_pred_all)]

    end.append(time.time())

print(f'xgb 평균 score : {sum(s_xgb)/len(s_xgb)}')
print(f'lgb 평균 score : {sum(s_lgb)/len(s_lgb)}')
print(f'deep 평균 score : {sum(s_deep)/len(s_deep)}')
#print(f'ext 평균 score : {sum(s_ext)/len(s_ext)}')
print(f'최종 평균 score : {sum(score_list)/len(score_list)}')

print(start)
print(middle)
print(end)
```
